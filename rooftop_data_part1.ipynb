```python
import requests

url = "https://yelp-reviews.p.rapidapi.com/business-search"

querystring = {"query":"rooftop bars","location":"Berlin","start":"0","yelp_domain":"yelp.com"}

headers = {
	"X-RapidAPI-Key": "",
	"X-RapidAPI-Host": "yelp-reviews.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
print(response.json()['data'])
```

```python
import pandas as pd
list=[]
data=[]
for i in range(len(response.json()['data'])):
    list=[response.json()['data'][i]['name'],response.json()['data'][i]['latitude'],response.json()['data'][i]['longitude'],response.json()['data'][i]['rating'],response.json()['data'][i]['review_count'],response.json()['data'][i]['price_range']]
    data.append(list)
df=pd.DataFrame(data)
df.columns=['name','latitude','longtitude','rating','review count','price range']
df['city']="Berlin"
df
```
